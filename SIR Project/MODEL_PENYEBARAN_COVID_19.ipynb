**A. KASUS I**

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

#total populadsi, N
N = 9176546
I0 = 2
R0 = 0
S0 = N - I0 - R0
beta, gamma = 0.2, 72621/(76992*30)
t = np.linspace(0, 365, 365)

def deriv(y, t, N, beta, gamma):
  S, I, R = y
  dSdt = -beta * S * I / N
  dIdt = beta * S * I/N - gamma * I
  dRdt = gamma * I
  return dSdt, dIdt, dRdt

y0 = S0, I0, R0
ret = odeint(deriv, y0, t, args=(N, beta, gamma))
S, I, R = ret.T

fig = plt.figure(facecolor ='w')
ax = fig.add_subplot(111, facecolor='#dddddd', axisbelow=True)
ax.plot(t, S/N, 'b', alpha=0.5, lw=2, label='susceptible')
ax.plot(t, I/N, 'r', alpha=0.5, lw=2, label='Infected')
ax.plot(t, R/N, 'g', alpha=0.5, lw=2, label='Recovered with immunity')
ax.set_xlabel('Time /days')
ax.set_ylabel('Number (1000s)')
ax.set_ylim(0, 1)
ax.yaxis.set_tick_params(length=0)
ax.xaxis.set_tick_params(length=0)
ax.grid(True, which='major', c='w', ls='-', lw=2)
legend = ax.legend()
legend.get_frame().set_alpha(0.5)
for spine in ('top', 'right', 'bottom', 'left'):
    ax.spines[spine].set_visible(False)
plt.title('Model Penyebaran COVID-19', color = 'red', fontsize = '20')
plt.show()

ax.plot(1)
fig = plt.figure(facecolor ='w')
ax = fig.add_subplot(111, facecolor='#dddddd', axisbelow=True)
ax.plot(t, S/N, 'b', alpha=0.5, lw=2, label='susceptible')
ax.set_xlabel('Time /days')
ax.set_ylabel('Number (1000s)')
ax.set_ylim(0, 1)
ax.yaxis.set_tick_params(length=0)
ax.xaxis.set_tick_params(length=0)
ax.grid(True, which='major', c='w', ls='-', lw=2)
legend = ax.legend()
legend.get_frame().set_alpha(0.5)
for spine in ('top', 'right', 'bottom', 'left'):
    ax.spines[spine].set_visible(False)
plt.title('A. Susceptible (Rentan Terinfeksi)', color = 'blue', fontsize = '12')
plt.legend()
plt.show()

ax.plot(2)
fig = plt.figure(facecolor ='w')
ax = fig.add_subplot(111, facecolor='#dddddd', axisbelow=True)
ax.plot(t, I/N, 'r', alpha=0.5, lw=2, label='Infected')
ax.set_xlabel('Time /days')
ax.set_ylabel('Number (1000s)')
ax.set_ylim(0, 1)
ax.yaxis.set_tick_params(length=0)
ax.xaxis.set_tick_params(length=0)
ax.grid(True, which='major', c='w', ls='-', lw=2)
legend = ax.legend()
legend.get_frame().set_alpha(0.5)
for spine in ('top', 'right', 'bottom', 'left'):
    ax.spines[spine].set_visible(False)
plt.title('B. Infected (Terinfeksi)', color = 'red', fontsize = '12')
plt.legend()
plt.show()

ax.plot(2)
fig = plt.figure(facecolor ='w')
ax = fig.add_subplot(111, facecolor='#dddddd', axisbelow=True)
ax.plot(t, R/N, 'g', alpha=0.5, lw=2, label='Recovered with immunity')
ax.set_xlabel('Time /days')
ax.set_ylabel('Number (1000s)')
ax.set_ylim(0, 1)
ax.yaxis.set_tick_params(length=0)
ax.xaxis.set_tick_params(length=0)
ax.grid(True, which='major', c='w', ls='-', lw=2)
legend = ax.legend()
legend.get_frame().set_alpha(0.5)
for spine in ('top', 'right', 'bottom', 'left'):
    ax.spines[spine].set_visible(False)
plt.title('C. Recovered (Sembuh))', color = 'green', fontsize = '12')
plt.legend()
plt.show()
```

**B. KASUS II**

Jika terdapat faktor penghambat penyebaran yaitu social distancing

misalkan rho sebagai penghambat penyebaran

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

#total populadsi, N
N = 9176546
I0 = 2
R0 = 0
S0 = N - I0 - R0
beta, gamma = 0.2, 72621/(76992*30)
rho = 0.5
t = np.linspace(0, 365, 365)

def deriv(y, t, N, beta, gamma, rho):
  S, I, R = y
  dSdt = -beta * rho * S * I / N
  dIdt = beta * rho * S * I/N - gamma * I
  dRdt = gamma * I
  return dSdt, dIdt, dRdt

y0 = S0, I0, R0
ret = odeint(deriv, y0, t, args=(N, beta, gamma, rho))
S, I, R = ret.T

fig = plt.figure(facecolor ='w')
ax = fig.add_subplot(111, facecolor='#dddddd', axisbelow=True)
ax.plot(t, S/N, 'b', alpha=0.5, lw=2, label='susceptible')
ax.plot(t, I/N, 'r', alpha=0.5, lw=2, label='Infected')
ax.plot(t, R/N, 'g', alpha=0.5, lw=2, label='Recovered with immunity')
ax.set_xlabel('Time /days')
ax.set_ylabel('Number (1000s)')
ax.set_ylim(0, 1)
ax.yaxis.set_tick_params(length=0)
ax.xaxis.set_tick_params(length=0)
ax.grid(True, which='major', c='w', ls='-', lw=2)
legend = ax.legend()
legend.get_frame().set_alpha(0.5)
for spine in ('top', 'right', 'bottom', 'left'):
    ax.spines[spine].set_visible(False)
plt.title('Model Penyebaran COVID-19 dengan Isolasi Mandiri', color = 'red', fontsize = '15')
plt.show()

ax.plot(1)
fig = plt.figure(facecolor ='w')
ax = fig.add_subplot(111, facecolor='#dddddd', axisbelow=True)
ax.plot(t, S/N, 'b', alpha=0.5, lw=2, label='susceptible')
ax.set_xlabel('Time /days')
ax.set_ylabel('Number (1000s)')
ax.set_ylim(0, 1)
ax.yaxis.set_tick_params(length=0)
ax.xaxis.set_tick_params(length=0)
ax.grid(True, which='major', c='w', ls='-', lw=2)
legend = ax.legend()
legend.get_frame().set_alpha(0.5)
for spine in ('top', 'right', 'bottom', 'left'):
    ax.spines[spine].set_visible(False)
plt.title('A. Susceptible (Rentan Terinfeksi)', color = 'blue', fontsize = '12')
plt.legend()
plt.show()

ax.plot(2)
fig = plt.figure(facecolor ='w')
ax = fig.add_subplot(111, facecolor='#dddddd', axisbelow=True)
ax.plot(t, I/N, 'r', alpha=0.5, lw=2, label='Infected')
ax.set_xlabel('Time /days')
ax.set_ylabel('Number (1000s)')
ax.set_ylim(0, 1)
ax.yaxis.set_tick_params(length=0)
ax.xaxis.set_tick_params(length=0)
ax.grid(True, which='major', c='w', ls='-', lw=2)
legend = ax.legend()
legend.get_frame().set_alpha(0.5)
for spine in ('top', 'right', 'bottom', 'left'):
    ax.spines[spine].set_visible(False)
plt.title('B. Infected (Terinfeksi)', color = 'red', fontsize = '12')
plt.legend()
plt.show()

ax.plot(2)
fig = plt.figure(facecolor ='w')
ax = fig.add_subplot(111, facecolor='#dddddd', axisbelow=True)
ax.plot(t, R/N, 'g', alpha=0.5, lw=2, label='Recovered with immunity')
ax.set_xlabel('Time /days')
ax.set_ylabel('Number (1000s)')
ax.set_ylim(0, 1)
ax.yaxis.set_tick_params(length=0)
ax.xaxis.set_tick_params(length=0)
ax.grid(True, which='major', c='w', ls='-', lw=2)
legend = ax.legend()
legend.get_frame().set_alpha(0.5)
for spine in ('top', 'right', 'bottom', 'left'):
    ax.spines[spine].set_visible(False)
plt.title('C. Recovered (Sembuh))', color = 'green', fontsize = '12')
plt.legend()
plt.show()
```
